```python
@application.route('/')

def hello_world():

    return 'Trendlinjer og konkordanser: /ngram?word=...., eller /konk?word=...&corpus=...&offset='

 

 

@application.route('/ner', methods=['POST'])

def named_entity():

    parameters = request.json

    text = parameters['text']

    dist = False

    if 'dist' in parameters:

        dist = parameters['dist']

    result =  ner(text, dist)

    return jsonify(result)

 

@application.route('/wordbags', methods=['POST'])

def wbags():

    parameters = request.json

    wordbags = parameters['wordbags']

    urns = parameters['urns']

    result = search_urns_wordbags_freq(wordbags, urns)

    return jsonify(result)

 

@application.route('/wordbags_para', methods=['POST'])

def wbags_para():

    parameters = request.json

    wordbags = parameters['wordbags']

    urns = parameters['urns']

    result = search_urns_para_wordbags_freq(wordbags, urns)

    return jsonify(result)

 

@application.route('/paragraphs')

def paragraphs():

    parameters = request.json

    paras = parameters['paragraphs']

    urn = parameters['urn']

    year = query(mods, "select year from biblio where urn = ?", (urn,))

    res = ""

    if int(year[0][0]) <= 1900:

        res = paras_from_text(urn, paras)

    return jsonify(res)

 

@application.route('/sentences', methods=['GET', 'POST'])

def api_sentences():

    urns = request.args.getlist('urns')

    num = xstr(request.args.get('num'), 10)

    result = get_corpus_sentences(urns, int(num))

    return jsonify(result)

 

@application.route('/book_aggregates', methods = ['POST'])

def aggregate_urns():

    parameters = request.json

   

    if not 'urns' in parameters:

        return jsonify('no urns')

   

    urnlist = parameters['urns']

    c = Counter()

   

    with sqlite3.connect(unigrams) as con:

        cur = con.cursor()

        for u in urnlist:

            cur.execute("select word, freq from unigram where urn = ?",(u,))

            c.update(dict(cur.fetchall()))

    return jsonify(c)

 

@application.route('/paradigm')

def paradigm_request():

    word = request.args.get('word')

    result = get_paradigms(word)

    return jsonify(result)

 

@application.route('/word_form')

def wordform_request():

    word = request.args.get('word')

    result = get_form(word)

    return jsonify(result)

 

 

@application.route('/word_lemma')

def wordlemma_request():

    word = request.args.get('word')

    result = get_lemma(word)

    return jsonify(result)

 

@application.route('/variant_form')

def variant_request():

    word = request.args.get('word')

    form = request.args.get('form')

    result = get_variant(word, form)

    return jsonify(result)

 

@application.route("/unigrams")

def unigram_request():

    word = request.args.get('word')

    period0 = request.args.get('period0')

    period1 = request.args.get('period1')

    ddk= request.args.get('ddk')

    topic= request.args.get('topic')

    gender= request.args.get('gender')

    lang = request.args.get('lang')

    trans = request.args.get('trans')

    publisher= request.args.get('publisher')

    media = request.args.get('media')

    name = request.args.get('name')

    if media == 'bok':

        res = unigram_book(

            word,

            (period0, period1),

            ddk = ddk, lang = lang,

            gender = gender,

            publisher = publisher,

            trans = trans, topic = topic)

    else:

        res = unigram_avis(word, (period0, period1))

       

    return jsonify(res)

 

@application.route("/bigrams")

def bigram_request():

    first = request.args.get('first')

    second = request.args.get('second')

    #if first == None or second == None:

        # exit here if not both are filled - can be changed ...

    #    return jsonify([])

    period0 = request.args.get('period0')

    period1 = request.args.get('period1')

    ddk= request.args.get('ddk')

    topic= request.args.get('topic')

    gender= request.args.get('gender')

    lang = request.args.get('lang')

    trans = request.args.get('trans')

    publisher= request.args.get('publisher')

    media = request.args.get('media')

    name = request.args.get('name')

    if media == 'bok':

        res = bigram_book(

            [first, second],

            (period0, period1),

            ddk = ddk, lang = lang,

            gender = gender, publisher = publisher,

            trans = trans, topic = topic)

    else:

        res = []

       

    return jsonify(res)

 

@application.route("/book_counts")

def book_count_request():

    period0 = request.args.get('period0')

    period1 = request.args.get('period1')

    res = year_counts_books((period0, period1))

    return jsonify(res)

 

@application.route('/ngram')

def ngram_request():

    word = request.args.get('word')

    corpus = request.args.get('corpus')

    cond_word = request.args.get('cond')

    topic = request.args.get('topic')

    ddk = request.args.get('ddk')

    name = request.args.get('paper_name')

    month = request.args.get('month')

    yearfrom = request.args.get('yearfrom')

    yearto = request.args.get('yearto')

    total = request.args.get('total')

    # set defaults

    if total == None:

        total = False

    elif total == 'yes' or total == True:

        total = True

    else:

        total = False

       

    if name == None:

        name = '#'

    if yearto == None and yearfrom == None:

        period = '#'

    else:

        if yearto == None:

            period = yearfrom

        elif yearfrom == None:

            period == yearto

        else:

            period = (yearfrom, yearto)

    if month == None:

        month == '#'

    if ddk == None:

        ddk = "#"

    if cond_word == None:

        cond_word = "#"

    if topic == None:

        topic = "#"

    if corpus == None:

        corpus='bok'

    if total == False:

        if corpus == 'avis':

            result = ngram_avis(word, avisnavn=name, cond_word=cond_word, period=period, month=month)

        elif corpus == 'avisnavn':

            result = avisnavn(word)

        else:

            result = ngram_bok(word, cond_word=cond_word, period=period, topic=topic, ddk=ddk)

    else:

        if corpus == 'avis':

            result = ngram_avis_tot(avisnavn=name, cond_word=cond_word, period=period, month=month)

        else:

            result = ngram_bok_tot( cond_word=cond_word, period=period, topic=topic, ddk=ddk)

   

    return jsonify(result)

 

 

 

@application.route('/konk')

def konk_request():

    word = request.args.get('word')

    corpus = xstr(request.args.get('corpus'),'bok')

    author=xstr(request.args.get('author'))

    title=xstr(request.args.get('title'))

    subtitle=xstr(request.args.get('subtitle'))

    lang=xstr(request.args.get('lang'))

    ddk = xstr(request.args.get('ddk'))

    subject = xstr(request.args.get('subject'))

    yearfrom = xstr(request.args.get('yearfrom'), 1820)

    yearto=xstr(request.args.get('yearto'), 2020)

    before = xstr(request.args.get('before'), 5)

    after = xstr(request.args.get('after'), 5)

    size = xstr(request.args.get('size'), 3)

    offset = xstr(request.args.get('offset'),0)

    gender = xstr(request.args.get('gender'),'')

   

    result = konk(word,

                corpus=corpus,

                offset=int(offset)*int(size),

                size=str(max(int(size), 4)),

                before=str(min(int(before),12)),

                after=str(min(int(after),12)),

                author=author,

                lang=lang,

                title=title,

                subject=subject,

                ddk=ddk,

                gender=gender,

                subtitle=subtitle,

                yearfrom=yearfrom,

                yearto=yearto)

 

    return result

 

@application.route('/collocation')

def collocation_request():

    word = request.args.get('word')

    corpus = xstr(request.args.get('corpus'),'bok')

    author=xstr(request.args.get('author'))

    title=xstr(request.args.get('title'),'%')

    subtitle=xstr(request.args.get('subtitle'))

    lang=xstr(request.args.get('lang'))

    ddk = xstr(request.args.get('ddk'))

    yearfrom = xstr(request.args.get('yearfrom'), 1820)

    yearto=xstr(request.args.get('yearto'), 2020)

    limit = xstr(request.args.get('limit'), 100)

    before = xstr(request.args.get('before'), 5)

    after = xstr(request.args.get('after'), 5)

    result = collocation(word,

                corpus=corpus,

                before=str(min(int(before),52)),

                after=str(min(int(after),52)),

                author=author,

                lang=lang,

                title=title,

                ddk=ddk,

                subtitle=subtitle,

                yearfrom=yearfrom,

                yearto=yearto,

                limit=limit)

    return result

 

@application.route('/wildcards')

def wildcard_search():

    books = "/disk1/bokhylla/digibok/book_words.db"

    wildword = request.args.get('word')

    factor= int(xstr(request.args.get('factor'), 2))

    freq_lim = int(xstr(request.args.get('freq_lim'), 50))

    limit= int(xstr(request.args.get('limit'),50))

    res = wildcard_hub(wildword, factor=factor, freq_lim=freq_lim, limit=limit)

    return jsonify(dict(res))

 

 

@application.route('/count', methods=['GET'])

def count_request():

    #word = request.args.get('word')

   

    corpus = xstr(request.args.get('corpus'),'bok')

    author=xstr(request.args.get('author'))

    title=xstr(request.args.get('title'))

    subtitle=xstr(request.args.get('subtitle'))

    lang=xstr(request.args.get('lang'),'nob')

    yearfrom = xstr(request.args.get('yearfrom'), 1820)

    yearto=xstr(request.args.get('yearto'), 2020)

    ddk = xstr(request.args.get('ddk'))

    result = count(

                corpus=corpus,

                author=author,

                lang=lang,

                title=title,

                subtitle=subtitle,

                ddk=ddk,

                yearfrom=yearfrom,

                yearto=yearto)

    #print(corpus, yearfrom)

    return jsonify(result)

 

 

 

@application.route('/freq', methods=['POST'])

def freq():

    parameters = request.json

    words = parameters['words']

    urn = parameters['urn']

    urndb = "/disk1/bokhylla/urn2db.db"

    #Take a list of words and return the frequencies

  

    

    db = query(urndb, "select dbfile from urn_db where urn =? ",(urn,))

    res = []

    sql = "select a.word, a.freq from unigram as a, freq_words.words as f where a.urn = ? and a.word = f.word"

    if db != []:

        db_file = db[0][0]

        with sqlite3.connect(db_file) as con:

            cur = con.cursor()

            cur.execute("attach '' as freq_words")

            cur.execute("create table freq_words.words (word varchar)")

            for w in words:

                cur.execute("insert into freq_words.words values ('{w}')".format(w=w))

        res = cur.execute(sql, (urn,)).fetchall()

    return jsonify(res)

 

@application.route('/book_count', methods = ['POST'])

def book_counts():

    parameters = request.json

    urns = parameters['urns']

    unigram = "/disk1/bokhylla/digibok/unigrams_bok.db"

    #Take a list of words and return the frequencies

    res = []

    for urn in urns:

        res += query(unigram, "select urn, count from urncount where urn = ?",(urn,))

    return jsonify(res)

 

@application.route('/totals', methods=['GET'])

def totals():

   

    db = "/disk1/bokhylla/digibok/book_words.db"

    parameters = request.json

    top = parameters['top']

    sql = "select word, tf from grand_total order by tf desc limit ?"

    res = query(db, sql, (top,), size=0)

    return jsonify(res)

 

@application.route('/word_frequencies', methods=['POST'])

def wfreq():

   

    unigram = "/disk1/bokhylla/digibok/book_words.db"

    params = request.json

    words = params['words']

    #Take a list of words and return the total frequencies   

    sql = "select a.word, a.tf from grand_total as a, freq_words.words as f where a.word = f.word"

    with sqlite3.connect(unigram) as con:

        cur = con.cursor()

        cur.execute("attach '' as freq_words")

        cur.execute("create table freq_words.words (word varchar)")

        for w in words:

            try:

                cur.execute("""insert into freq_words.words values ('{w}')""".format(w=w))

            except:

                True

        res = cur.execute(sql).fetchall()

    return jsonify(res)

 

@application.route('/meta', methods=['POST'])

def meta():

 

    mods = "/disk1/bokhylla/digibok/urn-biblio-mods.db"

    parameters = request.json

    urns = parameters['urn']

    res = []

    for urn in urns:

        res += konk_api.query(mods, "select urn, author, year, title, subtitle, publisher, genre, sesamid, lang, langtrans from biblio where urn = ?",(urn,), size=0)

    return jsonify(res)

 

@application.route('/word_graph', methods = ['POST'])

def make_word_graph():

    """ For making name graphs """

    parameters = request.json

    tokens = None

    tokenmap = None

    urn = None

    if 'tokens' in parameters:

        tokens = parameters['tokens']

    if 'urn' in parameters:

        urn = parameters['urn']

    if 'tokenmap' in parameters:

        tokenmap = parameters['tokenmap']

    if urn == None or tokens == None:

        res = []

    else:

        res = make_edges(urn, tokens, tokenmap)

    return jsonify(res)

 

@application.route('/word_counts', methods = ['POST'])

def make_word_counts():

    """ For making name counts - string counts """

    parameters = request.json

    tokens = None

    tokenmap = None

    urn = None

    if 'tokens' in parameters:

        tokens = parameters['tokens']

    if 'urn' in parameters:

        urn = parameters['urn']

    if 'tokenmap' in parameters:

        tokenmap = parameters['tokenmap']

    if urn == None or tokens == None:

        res = []

    else:

        res = count_strings(urn, tokens, tokenmap)

    return jsonify(res)

 

@application.route('/graph', methods=['POST'])

def make_graph():

    import sqlite3

    import pandas as pd

    import json

   

    

    urn_db = '/disk1/bokhylla/urn2db.db'

   

    parameters = request.json

    urn = parameters['urn']

    wordbag = parameters['words']

    db = query(urn_db, "select dbfile from urn_db where urn = ? ",(urn,))[0][0]

    print(db)

    with sqlite3.connect(db) as con:

        cur = con.cursor()

        cur.execute("attach '' as wordbag")

        cur.execute('create table wordbag.words (word varchar)')

        for word in wordbag:

            cur.execute("insert into wordbag.words values ('{u}')".format(u=word))

        cur.execute("""

        select  c.para, c.word, count(*)

        from ft as c

            inner join wordbag.words as b on b.word = c.word

        where c.urn = ?

        group by c.para, c.word

        """, (urn,))

        G = cur.fetchall()

        D = dict()

        for t in G:

            S = dict()

            for v in G:

                if t[0] == v[0]:

                    S[v[1]] = v[2]

            D[t[0]] = S

        G = pd.DataFrame(D).fillna(0)

        Gd = G.dot(G.transpose())

        J = json.loads(Gd.to_json())

        res = []

        for x in J:

            for y in J[x]:

                res.append((x,y,J[x][y]))

    return jsonify(res)

 

@application.route('/vekstdiagram', methods=['POST'])

def vekst_diagram():

    import sqlite3

    parameters = request.json

    urn = parameters['urn']

    window = 100

    if 'window' in parameters:

        window= parameters['window']

    pr = 100

    if 'pr' in parameters:

        pr = parameters['pr']

    if 'words' in parameters:

        words = parameters['words']

    else:

        words = ". , ! ? ; -".split()

       

    urndb = "/disk1/bokhylla/urn2db.db"

    #print(words)

    #parameters = request.json

    window = max(100, window)

    #pr = max(pr, 1000)

    db = konk_api.query(urndb, "select dbfile from urn_db where urn =? ",(urn,))

    if db != []:

        database = db[0][0]

        print(database)

        rowstart = konk_api.query(database, "select rowid from ft where urn = ? order by rowid limit 1",(urn,))

        rowend = konk_api.query(database, "select rowid from ft where urn = ? order by rowid desc limit 1",(urn,))

        if rowstart != [] and rowend !=[]:

            rowstart = rowstart[0][0]

            rowend = rowend[0][0] - window

            with sqlite3.connect(database) as con:

                cur = con.cursor()

                cur.execute("attach database '' as wordset")

                cur.execute("create table wordset.words (word varchar)")

                for w in words:

                    cur.execute("insert into wordset.words values (?)",(w,))

                count = []

                for row in range(rowstart, rowend, pr):

           

                    sql = """select count(b.word) from ft as a inner join wordset.words as b on b.word = a.word

                where a.urn = ? and a.rowid > ? and a.rowid <= ?"""

        #print(sql)

                    res = cur.execute(sql, (urn, row, row + window )).fetchall()

                    if res != []:

                        count.append(res[0][0])

                    if rowend - rowstart <= window:

                        break

    return jsonify(count)

 

@application.route("/sttr", methods=['GET'])

def sttr():

    """Compute type-token ratio from chunks and return average over chunks, where minimum chunk is 2000.

    The text is converted to lowercase, numbers and punctuation characters are removed."""

 

    params = request.json

    urn = params['urn']

    chunk = 5000

    if 'chunk' in params:

         chunk = params['chunk']

           

    urndb = "/disk1/bokhylla/urn2db.db"

    #Take a list of words and return the frequencies

   

    chunk = max(2000, chunk)

   

    db = query(urndb, "select dbfile from urn_db where urn =? ",(urn,))

    res = []

    count = []

    if db != []:

        database = db[0][0]

        #print(database)

        rowstart = query(database, "select rowid from ft where urn = ? order by rowid limit 1",(urn,))

        rowend = query(database, "select rowid from ft where urn = ? order by rowid desc limit 1",(urn,))

        if rowstart != [] and rowend !=[]:

            rowstart = rowstart[0][0]

            rowend = rowend[0][0]

            with sqlite3.connect(database) as con:

                cur = con.cursor()

                for row in range(rowstart, rowend, chunk):

                    sql = """select lower(word) as lw, count(*) from ft

                        where urn = ? and rowid > ? and rowid <= ? and not(length(word) = 1 and upper(word)=lower(word)) and abs(word) = 0 group by lw"""

                    res = cur.execute(sql, (urn, row, row + chunk)).fetchall()

                    if res != []:

                        count.append(res)

                       

    # compute the tt-score as average over chunks

    a = []

    for y in count:

        a.append(dict(y))

 

    z=pd.DataFrame(a).transpose()

 

    average_tt = (z.count()/z.sum()).mean()

   

    return jsonify(average_tt)

 

@application.route('/urnfreq', methods=['GET'])

def urn():

    urndb = "/disk1/bokhylla/urn2db.db"

    parameters = request.json

    urn = parameters['urn']

    keys = parameters.keys()

    top = 20

    if 'top' in keys:

        top = parameters['top']

    cutoff=5

    if 'cutoff' in keys:

        cutoff = parameters['cutoff']

    res = []

    db = konk_api.query(urndb, "select dbfile from urn_db where urn =? ",(urn,))

    if top > 0:

        sql = """select word, freq from unigram

            where urn = ? and freq > ? order by freq desc limit ?"""

        if db != []:

            res = konk_api.query(db[0][0], sql, (urn, cutoff, top), size=0)

    else:

        sql = """select word, freq from unigram

            where urn = ? and freq > ? order by freq desc"""

        if db != []:

            res = konk_api.query(db[0][0], sql, (urn, cutoff), size=0)

   

    return jsonify(res)

 

@application.route('/tingnavn', methods=['GET'])

def names():

    parameters = request.json

    urn = parameters['urn']

    ratio = 0.4

    if "ratio" in parameters:

        ratio = float(parameters['ratio'])

    frequency_threshold = 1

    if 'cutoff' in parameters:

        frequency_threshold = int(parameters['cutoff'])

 

    result = good_tokens(urn, ratio, frequency_threshold)

    return jsonify(result[0])

 

@application.route('/names', methods=['GET'])

def gnames():

    parameters = request.json

    urn = parameters['urn']

    ratio = 0.4

    if "ratio" in parameters:

        ratio = float(parameters['ratio'])

    frequency_threshold = 1

    if 'cutoff' in parameters:

        frequency_threshold = int(parameters['cutoff'])

 

    result = good_tokens(urn, ratio, frequency_threshold)

    return jsonify([[result[0]],list(result[1].items()), list(result[2].items()), list(result[3].items())])

 

 

@application.route('/avisfreq', methods=['GET'])

def fetchavis():

    urndb = "/disk1/bokhylla/alto_avis/urn_mods_avis.db"

    parameters = request.json

    keys = parameters.keys()

    navn = '%'

    if 'navn' in keys:

        navn = parameters['navn']

    yearfrom = 1800

    if 'yearfrom' in keys:

        yearfrom = parameters['yearfrom']

    yearto = '2018'

    if 'yearto' in keys:

        yearto = parameters['yearto']

    samplesize = 20

    if 'samplesize' in keys:

        samplesize= parameters['samplesize']

    top = 20

    if 'top' in keys:

        top = parameters['top']

    cutoff=5

    if 'cutoff' in keys:

        cutoff = parameters['cutoff']

    res = []

    db = konk_api.query(urndb, "select urn from avisdata where avisnavn like ? and year <= ? and year >= ? order by random() limit ?",(navn, yearto, yearfrom, samplesize), size=0)

    sql = """select word, freq from unigram where urn = ? and freq > ? order by freq desc"""

    #print(sql)

    aviser = []

    for x in db:

        avisfil = findavis(x[0])

        print(x[0], avisfil)

        aviser.append(konk_api.query(avisfil, sql, (x[0], cutoff), size=0))

    return jsonify(aviser)   

 

@application.route('/best_urn', methods = ['GET'])

def best_urns():

 

    params = request.json

    keys = params.keys()

    sql_condition = ""

    tuples = []

    if params == None:

        return "Need at least a word"

    if 'word' not in params:

        return "No word"

    else:

        word = params['word']

    keys = params

    if 'author' in keys:

        auth = params['author']

    else:

        auth = '%'

    select_statement = " distinct a.urn, a.author, a.title, a.year from biblio as a, unigram.unigram as f "

    sql_condition += " a.author like ? "

    tuples.append(auth)

    if 'gender' in keys:

        gender = params['gender']

        sql_condition += " and a.gender like ? "

        tuples.append(gender)

    if 'year' in keys:

        year = params['year']

    else:

        year = 1800

    sql_condition += " and a.year >= ?"

    tuples.append(year)

    neste = 10

    if 'neste' in keys:

        neste = params['neste']

    if 'next' in keys:

        neste = params['next']

    sql_condition += " and a.year <= ?"

    tuples.append(int(year) + int(neste))

    if 'title' in keys:

        title = params['title']

        sql_condition += " and a.title like ? "

        tuples.append(title)

    if 'ddk' in keys:

        ddk = params['ddk']

        sql_condition += " and c.class like ? and c.authority = 'ddc' and a.urn = c.urn "

        select_statement += ", classifications as c "

        tuples.append(ddk)

    if 'subject' in keys:

        subject = params['subject']

        sql_condition += " and s.topic like ? and s.urn = a.urn "

        select_statement += ", subjects as s "

        tuples.append(subject)

    if 'word' in keys:  # check code further down which depends on this

        #words = params['words'] set above

        sql_condition += " and a.urn = f.urn and f.word = ?  "

        tuples.append(word)

        #select_statement += ", words_meta.words as w, unigram as f "

        #words_are_metadata = True

    if 'lang' in keys:

        lang=params['lang']

        sql_condition += " and a.lang = ? "

        tuples.append(lang)

    if 'trans' in keys:

        trans = params['trans']

        sql_condition += " and a.langtrans = ? "

        tuples.append(trans)

    if 'limit' in keys:

        try:

            limit = int(params['limit'])

            sql_condition += " order by f.tfttf desc  limit ?"

            tuples.append(limit)

        except:

            True

    else:

        sql_condition += " order by f.tfttf desc limit 10"

 

    sql =  "select {select} where {condition}".format(select=select_statement, condition = sql_condition)

    print(sql, tuple(tuples))

    #return

    with sqlite3.connect(mods) as con:

        cur = con.cursor()

        cur.execute("attach database '' as words_meta")

        cur.execute("attach database '{unigram}' as unigram".format(unigram=unigrams))

            #cur.execute("""create table words_meta.words (word varchar)""")

            #for w in params['words']:

            #    cur.execute("""insert into words_meta.words values (?)""", (w,))

       

        cur.execute(sql, tuple(tuples))

        res = cur.fetchall()

 

    return jsonify(res)

 

@application.route('/urn', methods=['GET','POST'])

def urns():

   

 

    params = request.json

    keys = params.keys()

    sql_condition = ""

    tuples = []

    words_are_metadata = False

  

    if 'author' in keys:

        auth = params['author']

    else:

        auth = '%'

    select_statement = " distinct a.urn, a.author, a.title, a.year from biblio as a "

    sql_condition += " a.author like ? "

    tuples.append(auth)

    if 'gender' in keys:

        gender = params['gender']

        sql_condition += " and a.gender like ? "

        tuples.append(gender)

    if 'year' in keys:

        year = params['year']

    else:

        year = 1800

    sql_condition += " and a.year >= ?"

    tuples.append(year)

    neste = 10

    if 'neste' in keys:

        neste = params['neste']

    if 'next' in keys:

        neste = params['next']

    sql_condition += " and a.year <= ?"

    tuples.append(int(year) + int(neste))

    if 'title' in keys:

        title = params['title']

        sql_condition += " and a.title like ? "

        tuples.append(title)

    if 'subtitle' in keys:

        subtitle = params['subtitle']

        sql_condition += " and a.subtitle like ? "

        tuples.append(subtitle)

    if 'ddk' in keys:

        ddk = params['ddk']

        sql_condition += " and c.class like ? and c.authority = 'ddc' and a.urn = c.urn "

        select_statement += ", classifications as c"

        tuples.append(ddk)

    if 'subject' in keys:

        subject = params['subject']

        sql_condition += " and s.topic like ? and s.urn = a.urn "

        select_statement += ", subjects as s"

        tuples.append(subject)

    if 'words' in keys:  # check code further down which depends on this

        words = params['words']

        sql_condition += " and w.word = f.word and f.urn = a.urn "

        select_statement += ", words_meta.words as w, unigram as f "

        words_are_metadata = True

    if 'lang' in keys:

        lang=params['lang']

        sql_condition += " and a.lang = ? "

        tuples.append(lang)

    if 'trans' in keys:

        trans = params['trans']

        sql_condition += " and a.langtrans = ? "

        tuples.append(trans)

    if 'limit' in keys:

        try:

            limit = int(params['limit'])

            sql_condition += " order by random()  limit ?"

            tuples.append(limit)

        except:

            True

    else:

        sql_condition += "order by random() limit 10"

 

    sql =  "select {select} where {condition}".format(select=select_statement, condition = sql_condition)

    with sqlite3.connect(mods) as con:

        cur = con.cursor()

        if words_are_metadata:

            cur.execute("attach database '' as words_meta")

            cur.execute("attach database '{unigram}' as unigram".format(unigram=unigrams))

            cur.execute("""create table words_meta.words (word varchar)""")

            for w in params['words']:

                cur.execute("""insert into words_meta.words values (?)""", (w,))

        cur.execute(sql, tuple(tuples))

        res = cur.fetchall()

    return jsonify(res)

 

@application.route('/refineurn', methods=['GET', 'POST'])

def refine_urns():

 

   

    params = request.json

    keys = params.keys()

    sql_condition = ""

    tuples = []

    words_are_metadata = False

    if 'author' in keys:

        auth = params['author']

    else:

        auth = '%'

    select_statement = " distinct a.urn, a.author, a.title, a.year from biblio as a, urn_list.urns as urns"

    sql_condition += " a.author like ? and a.urn = urns.urn "

    tuples.append(auth)

    if 'gender' in keys:

        gender = params['gender']

        sql_condition += " and a.gender like ? "

        tuples.append(gender)

    if 'year' in keys:

       year = params['year']

    else:

        year = 1800

    sql_condition += " and a.year >= ?"

    tuples.append(year)

    neste = 10

    if 'neste' in keys:

        neste = params['neste']

    if 'next' in keys:

        neste = params['next']

    sql_condition += " and a.year <= ?"

    tuples.append(int(year) + int(neste))

    if 'title' in keys:

        title = params['title']

        sql_condition += " and a.title like ? "

        tuples.append(title)

    if 'ddk' in keys:

        ddk = params['ddk']

        sql_condition += " and c.class like ? and c.authority = 'ddc' and a.urn = c.urn "

        select_statement += ", classifications as c"

        tuples.append(ddk)

    if 'subject' in keys:

        subject = params['subject']

        sql_condition += " and s.topic like ? and s.urn = a.urn "

        select_statement += ", subjects as s"

        tuples.append(subject)

    if 'words' in keys:  # check code further down which depends on this

        words = params['words']

        if words != []:

            sql_condition += " and w.word = f.word and f.urn = a.urn "

            select_statement += ", words_meta.words as w, unigram as f"

            words_are_metadata = True

    else:

        params['words'] = []

    if 'lang' in keys:

        lang=params['lang']

        sql_condition += " and a.lang = ? "

        tuples.append(lang)

    if 'trans' in keys:

        trans = params['trans']

        sql_condition += " and a.langtrans = ? "

        tuples.append(trans)

    if 'limit' in keys:

        try:

            limit = int(params['limit'])

            sql_condition += " order by random()  limit ?"

            tuples.append(limit)

        except:

            True

    else:

        sql_condition += "order by random() limit 10"

 

    sql =  "select {select} where {condition}".format(select=select_statement, condition = sql_condition)

    with sqlite3.connect(mods) as con:

        cur = con.cursor()

        cur.execute("attach database '{unigram}' as unigram".format(unigram=unigrams))

        cur.execute("attach database '' as urn_list")

        cur.execute("create table urn_list.urns (urn int)")

        cur.execute("attach database '' as words_meta")

        cur.execute("""create table words_meta.words (word varchar)""")

        for u in params['urns']:

            cur.execute("""insert into urn_list.urns values (?)""", (u,))

        for w in params['words']:

            cur.execute("""insert into words_meta.words values (?)""", (w,))

        cur.execute(sql, tuple(tuples))

        res = cur.fetchall()

    #print(sql, tuple(tuples))

    return jsonify(res)

       

 

 

def findavis(num):

    x = int(num / 20000)

 

    return "/disk1/bokhylla/alto_avis/alto_avis_{s}_{e}.db".format(s = x * 20000 + 1, e = (x + 1)*20000 )

 

 

@application.route('/urnkonk', methods=[ 'POST'])

def urn_konk_request():

    parameters = request.json

    word = parameters['word']

    urns = parameters['urns']

    param_keys = parameters.keys()

    if 'corpus' in param_keys:

        corpus = parameters['corpus']

    else:

        corpus = 'bok'

    if 'size' in param_keys:

        size = parameters['size']

    else:

        size = 5

    if 'before' in param_keys:

        before = parameters['before']

    else:

        before = 5

    if 'after' in param_keys:

        after = parameters['after']

    else:

        after = 5

    offset =  0

    result = konk_urn(word,

                corpus=corpus,

                offset=0,

                size=str(max(int(size), 4)),

                before=str(min(int(before),12)),

                after=str(min(int(after),12)),

                urns=urns)

    return result

 

@application.route('/urncoll', methods=[ 'POST'])

def urn_coll_request():

    parameters = request.json

    word = parameters['word']

    urns = parameters['urns']

    param_keys = parameters.keys()

    # number of hits pr. urn

    if 'limit' in param_keys:

        limit = parameters['limit']

    else:

        limit = 5

    if 'before' in param_keys:

        before = parameters['before']

    else:

        before = 5

    if 'after' in param_keys:

        after = parameters['after']

    else:

        after = 5

    result = urn_coll_json_book(word,

                limit=limit,

                before=before,

                after=after,

                urns=urns)

    return jsonify(result)

 

@application.route('/urncolldist', methods=[ 'POST'])

def urn_coll_dist_request():

    parameters = request.json

    word = parameters['word']

    urns = parameters['urns']

    param_keys = parameters.keys()

    # number of hits pr. urn

    if 'limit' in param_keys:

        limit = parameters['limit']

    else:

        limit = 5

    if 'before' in param_keys:

        before = parameters['before']

    else:

        before = 5

    if 'after' in param_keys:

        after = parameters['after']

    else:

        after = 5

    result = urn_coll_dist_json_book(word,

                limit=limit,

                before=before,

                after=after,

                urns=urns)

    return jsonify(result)

   
```

```python
def wordbags(words, urns):
    url = 'https://api.nb.no/ngram/wordbags'
    if isinstance(words, str):
        return
    if isinstance(words, list):
        words = {w:[w] for w in words}
    r = requests.post(url, json={'wordbags':words, 'urns':urns})
    return r.json()
```

```python
def word_frequencies(words):
    url = 'https://api.nb.no/ngram/word_frequencies'
    if isinstance(words, str):
        words = words.split()
    r = requests.post(url, json={'words':words})
    return r.json()
```
